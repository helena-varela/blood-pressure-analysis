**Bibliotecas utilizadas**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Coleta de dados**

```python
# Coleta os dados de uma semana e armazena em listas
systolic = []
diastolic = []

for i in range(1, 8):
  print(f"Day {i}")
  s = int(input("Systolic pressure: "))
  d= int(input("Diastolic pressure: "))
  systolic.append(s)
  diastolic.append(d)
```

Plotagem do Gráfico

```python
x = np.arange(1, 8)

# Graph Plotting
plt.plot(x, systolic, 'r-o', label = "Systolic (mmHg)", linewidth=2, markersize=8 )
plt.plot(x, diastolic, 'b-o', label = "Diastolic (mmHg)", linewidth=2, markersize=8 )

# Graph Minimum and Maximum
min_val = min(min(systolic), min(diastolic)) - 5
max_val = max(max(systolic), max(diastolic)) + 5

plt.ylim(
    min(min(systolic), min(diastolic)) - 5,
    max(max(systolic), max(diastolic)) + 5
)

# Add the point values
for i in range(len(x)):
    plt.annotate(f'{systolic[i]}', (x[i], systolic[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)
    plt.annotate(f'{diastolic[i]}', (x[i], diastolic[i]), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8)

# Visual elements
plt.xticks(x, ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.yticks(np.arange(min_val, max_val + 1, 5))
plt.xlabel('Days of the week')
plt.ylabel('Blood pressure (mmHg)')
plt.title('Blood Pressure Variation')
plt.legend(loc='center')
plt.grid(True)
plt.show()
```
